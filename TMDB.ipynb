# Demographic Filtering

 In demographic filtering, we will suggest a movie based on the rating, score, popularity of the movie which is more of generalized way of giving suggestions, this is also an disadvantage of this type of suggestion. Because every user will have their own preference, not all users will prefer highly rated movies. so this type of suggestions cannot identify specific users interest.
 
 Using below formula, we will calculate the weighted rating and suggest the user which have high rating. we can use vote average to pick the movies but the shortcoming of this approach is in some cases movies will have high 'vote average' and low 'vote count', so movies with high 'vote count' and average 'vote average' will not be picked up. this will result in poor movie suggestions. so we compute WR using below formula


 weighted rating (WR) = (v ÷ (v+m)) × R + (m ÷ (v+m)) × C
 
 Where:

- R = average for the movie (mean) = (Rating)
- v = number of votes for the movie = (votes)
- m = minimum votes required to be listed in the Top 250 (currently 25,000)
- C = the mean vote across the whole report

```python
#We already have v(vote_count) and R (vote_average) and C can be calculated as
```

```python
c=df['vote_average'].mean()
c
```

```python
#for 'm' we can take 90 percentile of all values
```

```python
m=df['vote_count'].quantile(0.9)
m
```

```python
#we are taking top rated movies having count more than 'm'
top_movies=df.copy().loc[df['vote_count']>m]
top_movies.shape
```

```python
top_movies.head(1)
```

```python
v=top_movies['vote_count']
r=top_movies['vote_average']
```

```python
top_movies['score']=(v/(v+m) * r) + (m/(m+v) * c)
```

```python
#we are sorting in descending order to get the list in rating wise
top_movies=top_movies.sort_values('score',ascending=False)
```

```python
# here is our top rated movies which have considerable number of votes
top_movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```

## Plot based suggestions

In this type, we will get the input movie from user and will give the list of movies having similar plot as of user's input movie

Machine cannot compare the plot of two movies since the plot is textual data. To make it machine understandable we will convert the plot text data into numerical values using tfid which will map a numerical value to each word. so that the machine will be able to find a pattern by using cosine similarty function. cosine similarity score is between 0 to 1. Nearer to 1 shows that two movies are similar in plot wise.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
```

```python
#overview column has 3 null values. so we are filling it with spaces
df2.info()
```

```python
df2['overview'] = df2['overview'].fillna('')
```

```python
tfidf_matrix = tfidf.fit_transform(df2['overview'])
```

```python
tfidf_matrix.shape
```

```python


from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
cosine_sim.shape
```

```python
#we are creating a dataframe frame of cosine similarity values obtained by comparing each movie with every other movie
# we can see that the value is 1 when a movie is compare to itself
csdf=pd.DataFrame(cosine_sim,index=df2['title'],columns=df2['title'])
csdf.iloc[:7,:7]
```

```python
#omiting the first title since the title name is same 
#below are the list of movies which are similar to the movie given by the user
csdf['The Avengers'].sort_values(ascending=False)[1:11]
```

```python
csdf['The Dark Knight Rises'].sort_values(ascending=False)[1:11]
```

```python
#Picking the top 10 movies similar to each movie and creating a dataframe
titles=list(csdf.columns)
plot_dict=dict()
for i in range(len(titles)):
    plot_dict[titles[i]]=list(csdf.iloc[i].sort_values(ascending=False)[1:11].index)
```

```python
plot_df=pd.DataFrame(plot_dict)
plot_df.to_csv('plot.csv',index=False)
```

# Recommandation based on Genre, Cast, Crew, Keywords

By this we can recommand users based on Genre, Cast, Crew, Keywords. Here also same approach as plot based recommandation. we will extract the four columns and join them to create a single column. then we can convert the textual data to numerical data. cosine similarity values will be computed and based on the scores the movies will be suggested.

```python
dft=df.copy()
```

```python
features = ['cast', 'crew', 'keywords', 'genres']
for i in features:
    print(type(dft[i][0]))
```

```python
#since the cast, crew, keywords, genres are type string, we cant parse the directionaries present inside the string
#so we are using literal_eval to parse the stringified features into their corresponding python objects
from ast import literal_eval

for feature in features:
    dft[feature] = dft[feature].apply(literal_eval)
```

```python
def director(x):
    for i in x:
        if(i['job']=='Director'):
            return i['name']
    return np.nan    

def extract(x):
    temp=[]
    for i in x:
        if len(temp)>2:
            break
        else:
            temp.append(i['name'])
    return temp
```

```python
dfdir = pd.DataFrame() 
dfdir['Title']=dft['title']
dfdir['Cast']=dft['cast'].apply(extract)
dfdir['Director']=dft['crew'].apply(director)
dfdir['Keywords']=dft['keywords'].apply(extract)
dfdir['Genres']=dft['genres'].apply(extract)
```

```python
#since name of cast, crew, keywords have spaces vectorization will result in 2 vectors for same name 
# eg: Tom cruise and Tom Hardy will be little bit similar because of 'Tom' in their name
#so we are going to combine the first and last name to ger single vector
def merge(x):
    temp=[]
#cast, crew, keywords are list,so they satisfy if
    if isinstance(x,list):
        for i in x:
            temp.append(str.lower(i.replace(" ","")))
        return temp
#director is string so will come into else and some director name have NaN values
    else:
        if isinstance(x,str):
            return str.lower(x.replace(" ",""))
        else:
            return ''
#we are combining cast, crew, keywords,director        
def extract(x):
    return ' '.join(x['Keywords']) + ' ' + ' '.join(x['Cast']) + ' ' + x['Director'] + ' ' + ' '.join(x['Genres'])
```

```python
features = ['Cast','Director','Keywords', 'Genres']
for feature in features:
    dfdir[feature] = dfdir[feature].apply(merge)
dfdir['extract']=dfdir.apply(extract,axis=1)
dfdir.head(1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(dfdir['extract'])
count_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
cosine_sim2.shape
```

```python
#we are creating a dataframe frame of cosine similarity values obtained by comparing each movie with every other movie
# we can see that the value is 1 when a movie is compare to itself
csdf2=pd.DataFrame(cosine_sim2,index=df2['title'],columns=df2['title'])
csdf2.iloc[:7,:7]
```

```python
#omiting the first title since the title name is same 
#below are the list of movies which are similar to the movie given by the user
csdf2['The Avengers'].sort_values(ascending=False)[1:11]
```

```python
#omiting the first title since the title name is same 
#below are the list of movies which are similar to the movie given by the user
csdf2['The Dark Knight Rises'].sort_values(ascending=False)[1:11]
```

```python
csdf2['The Godfather'].sort_values(ascending=False)[1:11]
```

```python
#Picking the top 10 movies similar to each movie and creating a dataframe
titles2=list(csdf2.columns)
plot_dict2=dict()
for i in range(len(titles2)):
    plot_dict2[titles2[i]]=list(csdf2.iloc[i].sort_values(ascending=False)[1:11].index)
```
